### Problem 1.

The directories at http://www.govtrack.us/data/congress/113/votes/2013 contain JSON information about every vote cast for the current (113th)  Congress. Subdirectories beginning with "S" correspond to Senate votes, while subdirectories beginning with "H" correspond to House votes.

Write two functions: one that downloads and parses a single Senate vote page given the vote number, and another that repeatedly calls this function to build a full collection of Senate votes from the 113th Congress.

```python
"""
Function
--------
get_senate_vote

Scrapes a single JSON page for a particular Senate vote, given by the vote number

Parameters
----------
vote : int
   The vote number to fetch
   
Returns
-------
vote : dict
   The JSON-decoded dictionary for that vote
   
Examples
--------
>>> get_senate_vote(12)['bill']
{u'congress': 113,
 u'number': 325,
 u'title': u'A bill to ensure the complete and timely payment of the obligations of the United States Government until May 19, 2013, and for other purposes.',
 u'type': u'hr'}
"""
#your code here
def get_senate_vote(vote):
    url = 'https://www.govtrack.us/data/congress/113/votes/2013/s%s/data.json'%vote
    data = requests.get(url).text
    return json.loads(data)
```

```python
"""
Function
--------
get_all_votes

Scrapes all the Senate votes from http://www.govtrack.us/data/congress/113/votes/2013,
and returns a list of dicts

Parameters
-----------
None

Returns
--------
votes : list of dicts
    List of JSON-parsed dicts for each senate vote
"""
#Your code here
def get_all_votes():
    dom = web.Element(requests.get('http://www.govtrack.us/data/congress/113/votes/2013').text)
    votes = [] #list of JSON_parsed dicts
    
    #find each vote number and get the associated dictionary
    for a in dom('a'):
        if a.content.startswith('s'):
            vote = a.content[1:-1] #remove the 's' and '/'
            votes.append(get_senate_vote(vote))

    return votes
```

### Problem 2

Now, turn these data into a NetworkX graph, according to the spec below. For details on using NetworkX, consult the lab materials for November 1, as well as the [NetworkX documentation](http://networkx.github.io/).

```python
"""
Function
--------
vote_graph

Parameters
----------
data : list of dicts
    The vote database returned from get_vote_data

Returns
-------
graph : NetworkX Graph object, with the following properties
    1. Each node in the graph is labeled using the `display_name` of a Senator (e.g., 'Lee (R-UT)')
    2. Each node has a `color` attribute set to 'r' for Republicans, 
       'b' for Democrats, and 'k' for Independent/other parties.
    3. The edges between two nodes are weighted by the number of 
       times two senators have cast the same Yea or Nay vote
    4. Each edge also has a `difference` attribute, which is set to `1 / weight`.

Examples
--------
>>> graph = vote_graph(vote_data)
>>> graph.node['Lee (R-UT)']
{'color': 'r'}  # attributes for this senator
>>> len(graph['Lee (R-UT)']) # connections to other senators
101
>>> graph['Lee (R-UT)']['Baldwin (D-WI)']  # edge relationship between Lee and Baldwin
{'difference': 0.02, 'weight': 50}
"""
#Your code here

from collections import defaultdict
from operator import itemgetter
import itertools

def vote_graph(data):
    graph = nx.Graph()
    edges = []
    party_to_color = {'R':'r','D':'b', 'I':'k'}
    
    for bill in data:
        #create a list of nodes in the Nay cateogry and a list of nodes in the Yae category 
        #each node is of the format (display name, color based on party affiliation)
        #note: I had try-except statements to see if there are party affiliations other than R, D, and I, 
        #      but I did not find any, so I removed the try-except statements
        Nayers = [(Nayer['display_name'], dict(color=party_to_color[Nayer['party']])) for Nayer in bill['votes']['Nay']]
        Yeaers = [(Yeaer['display_name'], dict(color=party_to_color[Yeaer['party']])) for Yeaer in bill['votes']['Yea']]
        
        #add nodes to graph
        graph.add_nodes_from(Nayers)
        graph.add_nodes_from(Yeaers)
        
        #create a list of edges based on common Yea or Nay votes in any bill
        #itertools.combinations creates 2-length tuples, in sorted order, with no repeated elements
        #so there will not be two edges with the order of the nodes flipped (these two edges are the same in the undirected graph)
        #add the edges found from this bill to the total list of edges
        edges += list(itertools.combinations(map(itemgetter(0),Nayers), 2))
        edges += list(itertools.combinations(map(itemgetter(0),Yeaers), 2))
    
    edges.sort() #itertools.groupby below works on only sorted data 
    
    edges_by_weight = defaultdict(list)
    for edge, group in itertools.groupby(edges):
        edges_by_weight[len(list(group))].append(edge)
    
    #add edges to graph
    for w in edges_by_weight.keys():
        graph.add_edges_from(edges_by_weight[w], weight=w, difference=1.0/w)
        
    return graph
```

```python
votes = vote_graph(vote_data)    
```

### How (and how not) to visualize networks

Network plots often *look* impressive, but creating sensible network plots is tricky. From Ben Fry, the author of the Processing program:
<blockquote>
Usually a graph layout isn’t the best option for data sets larger than a few dozen nodes. You’re most likely to wind up with enormous spider webs or balls of string, and the mess seen so far is more often the case than not. Graphs can be a powerful way to represent relationships between data, but they are also a very abstract concept, which means that they run the danger of meaning something only to the creator of the graph. Often, simply showing the structure of the data says very little about what it actually means, even though it’s a perfectly accurate means of representing the data. Everything looks like a graph, but almost nothing should ever be drawn as one.
</blockquote>

Let's look at bad and better ways of visualizing the senate vote network.

```python
#your code here

#I propose a variant of closeness centrality - 1/average distance to all the nodes affiliated with the other party
#this measure does not apply to the nodes that are neither Democratic or Republican

def get_all_party_member(graph, party):
    """Given a network graph and party name, return a list of all nodes affiliated with the given party in the network 
    """
    
    return [node for node in graph if graph.node[node]['color']==party]

def closeness_centrality_variant(graph, distance='weight'):
    """Given a network graph and an edge attribute name as edge distance, return a dictionary 
    with node name as the key and the variant of closeness centrality as the value.
    
    For Democratic or Republic nodes, the variant of closeness centrality is defined to be 
    1/average distance to all nodes affiliated with the other party
    
    For Independent/other parties nodes, the variant of closeness centrality is set to be 0
    (to indicate their irrlevance in bipartisanship)
    """
    res = {}
    democrats = get_all_party_member(graph, 'b')
    republicans = get_all_party_member(graph, 'r')
        
    for node in graph:
        if graph.node[node]['color'] == 'r': #this node is Republican
            node_list = democrats + [node]
            subnet = graph.subgraph(node_list)
            res[node] = nx.closeness_centrality(subnet, u=node, distance=distance)
        elif graph.node[node]['color'] == 'b': #this node is Democratic
            node_list = republicans + [node]
            subnet = graph.subgraph(node_list)
            res[node] = nx.closeness_centrality(subnet, u=node, distance=distance)
        elif graph.node[node]['color'] == 'k': #this node is independent/other parties
            res[node] = 0
            
    return res
            
    
centrality_variant = closeness_centrality_variant(votes, distance='difference')

#repeat the network visualization codes as above
#note changes in the parameters of nx.spring_layout

#visualize the closeness_centrality score by NODE SIZE
#the centrality scores are raised to power 3 to make the node size different more noticeable
size = np.power(np.array(centrality_variant.values()), 3)/200.0

#this makes sure draw_spring results are the same at each call
np.random.seed(1)  

color = [votes.node[senator]['color'] for senator in votes.nodes()]

#determine position of each node using a spring layout
pos = nx.spring_layout(votes, k=0.3, iterations=20)

#plot the edges
nx.draw_networkx_edges(votes, pos, alpha = .05)

#plot the nodes
nx.draw_networkx_nodes(votes, pos, node_size=size, node_color=color)

#draw the labels
lbls = nx.draw_networkx_labels(votes, pos, alpha=5, font_size=8)

#coordinate information is meaningless here, so let's remove it
plt.xticks([])
plt.yticks([])
remove_border(left=False, bottom=False)
plt.show()


#visualize the variant of closeness_centrality score by a bar graph
data = [(name, centrality_variant[name], votes.node[name]['color']) for name in centrality.keys()]
data = sorted(data, key=itemgetter(1), reverse=True)
width=0.6
plt.bar(left=range(len(data)), height=map(itemgetter(1),data), width=width, color=map(itemgetter(2),data))
plt.ylabel('Variant of Closeness Centrality')
plt.title('Distribution of Variant of Closeness Centrality')
plt.xticks(np.arange(len(data))+width/2, map(itemgetter(0),data), rotation=90, size=6)
plt.xlim([0, len(data)])
remove_border()
plt.show()
```

## Leadership in the Senate

There are many metrics to quantify the leadership in the Senate.

  1. Senate leaders sponsor and co-sponsor lots of bills
  2. Leaders sit on many committees, as well as more important committees
  3. Leaders usually have been in office for a long time
  
Another approach uses the philosophy behind how Google ranks search results. The core idea behind Google's PageRank algorithm is:

1. A "good" website (i.e. one to rank highly in search results) is linked to by many other websites
2. A link found on a "good" website is more important than a link found on a "bad" website

The PageRank algorithm thus assigns scores to nodes in a graph based on how many neighbors a node has, as well as the score of those neighbors.

This technique can be adapted to rank Senate leadership. Here, nodes correspond to Senators, and edges correspond to a senator co-sponsoring a bill sponsored by another Senator. The weight of each edge from node A to B is the number of times Senator A has co-sponsored a bill whose primary sponsor is Senator B. If you interpret the PageRank scores of such a network to indicate Senate leadership, you are then assuming:

1. Leaders sponsor more bills
1. Leaders attract co-sponsorship from other leaders

### Problem 7

Govtrack stores information about each Senate bill in the current congress at http://www.govtrack.us/data/congress/113/bills/s/. As in problem 1, write two functions to scrape these data -- the first function downloads a single bill, and the second function calls the first to loop over all bills.

```python
"""
Function
--------
get_senate_bill

Scrape the bill data from a single JSON page, given the bill number

Parameters
-----------
bill : int
   Bill number to fetch
   
Returns
-------
A dict, parsed from the JSON

Examples
--------
>>> bill = get_senate_bill(10)
>>> bill['sponsor']
{u'district': None,
 u'name': u'Reid, Harry',
 u'state': u'NV',
 u'thomas_id': u'00952',
 u'title': u'Sen',
 u'type': u'person'}
>>> bill['short_title']
u'Agriculture Reform, Food, and Jobs Act of 2013'
"""
#your code here
def get_senate_bill(bill):
    url = 'https://www.govtrack.us/data/congress/113/bills/s/s%s/data.json'%bill
    data = requests.get(url).text
    return json.loads(data)
```

```python
"""
Function
--------
get_all_bills

Scrape all Senate bills at http://www.govtrack.us/data/congress/113/bills/s

Parameters
----------
None

Returns
-------
A list of dicts, one for each bill
"""
#your code here
def get_all_bills():
    dom = web.Element(requests.get('https://www.govtrack.us/data/congress/113/bills/s/').text)
    bills = [] #list of JSON_parsed dicts
    
    #find each bill number and get the associated dictionary
    for a in dom('a'):
        bill = a.content[1:-1] #remove the 's' and '/'
        if bill.isdigit():
            bills.append(get_senate_bill(bill))

    return bills
```

```python
bill_list = get_all_bills()
```

### Problem 8

Write a function to builded a Directed Graph (DiGraph) from these data, according to the following spec:

```python
"""
Function
--------
bill_graph

Turn the bill graph data into a NetworkX Digraph

Parameters
----------
data : list of dicts
    The data returned from get_all_bills

Returns
-------
graph : A NetworkX DiGraph, with the following properties
    * Each node is a senator. For a label, use the 'name' field 
      from the 'sponsor' and 'cosponsors' dict items
    * Each edge from A to B is assigned a weight equal to how many 
      bills are sponsored by B and co-sponsored by A
"""
#Your code here
def bill_graph(data):
    graph = nx.DiGraph()
    nodes = set()
    edges = defaultdict(int) #a dictionary with edge as key and weight as value
    
    for bill in data:
        sponsor = bill['sponsor']['name']
        nodes.add(sponsor)
        
        for cosponsor in bill['cosponsors']:
            cosponsor = cosponsor['name']
            nodes.add(cosponsor)
            edges[(cosponsor, sponsor)] += 1
    
    #add edges to graph
    for node1, node2 in edges:
        graph.add_edge(node1, node2, weight=edges[(node1, node2)])        
        
    return graph
```

```python
bills = bill_graph(bill_list)
```

The top five senators with the highest PageRank scores have node degrees ranked in the top quartile.
