## I. Background, Motivation, and Relevant Literature

Recent years have seen rising importance of network sciences in both our daily lives and the scholastic world (e.g. the image above shows a citation network, which is one way in which network science has been applied to study the development of knowledge). The emergence of the Internet and social network sites has massively changed the way in which humans connect and interact. Realizing the role that a smaller world and increasing connectedness play in the modern world, scholars in various disciplines directed intellectual curiosity and efforts to studying networks. They frequently draw from works of their colleagues, as well as intellectuals from other fields. Since the late 1950s, the number of network-related JSTOR articles published per year has grown exponentially. It reached unprecedented levels of over 1200 publications per year in 2010.

What is more interesting about network-related topics beyond the sheer magnitude of scholarly interest in the topic is the interconnectedness between different disciplines studying the topic. As evidenced by the appearance of network-related topics in many areas of study, the network science is a highly interdisciplinary subject. It draws on theories and methods from mathematics, physics, computer science, statistics, sociology, and so on. As researchers from different disciplines collaborate and cite each others’ work, the topics studied expand, and ideas spread across fields to generate new results.

Our work is inspired by a famous sociology article by Mark Granovetter (1973) - “The Strength of Weak Ties” published in the American Journal of Sociology - and the follow-up discussion between Mark Granovetter and Herbert Gans on the interesting sociological issues of structure, agency, and culture in community networks. However, by the time Duncan Watt reviewed the development of network science in the Annual Review of Sociology in 2004, he primarily focuses on the progress of network studies in mathematics and physics, and he overlooks the sociological issues of networks. According to Harvard Sociologist Professor Robert Sampson, network science started out from sociology, but it has now mostly moved out of the reach of sociology and evolved into its own field. In this project, we set out to explore the degree of collaboration across disciplines and the prominence of each discipline within network science.

The goal of our project is to trace the evolution of network-related topics in academia and to quantify and visualize the degree to which network research cuts across various disciplines. Three major tasks are sought to be completed. First, we group network-related topics by discipline and explore the question of which disciplines are the most central in network science. Second, we investigate the citation relationships of articles between disciplines. Third, we fit our data on the number of articles and citation relationships across years into a network diffusion model, and probe into the mechanism through which network-related topics spread through academic publications.

### Web Scraping --- Getting Network Articles from JSTOR Database

First, here are our helper functions. Sometimes the article information contains html tags, so we found this nice function strip_tags that can remove the unneeded html tags from this website http://stackoverflow.com/questions/753052/strip-html-from-strings-in-python

We also wrote the funciton get_web to optimize web scraping.

````python
class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    """
    Remove html tags in the given string and return the edited string
    """
    
    s = MLStripper()
    s.feed(html)
    return s.get_data()
````

````python
def get_web(url):
    """
    Given the url of a web page, return its web content (an object in the web.Element class)
    """
    
    #notes by Xueshan:
    # I found it takes 3+ seconds to request the web.Element, but only less than 1 second to get cites or abstract.
    # To get both citations and abstract in a minimum time, I took out the request web.Element part here. 

    dom = web.Element(requests.get(url).text)
    return dom
````

Next are our main functions to get the articles from online. We basically created a separate function to scrape each of the elements we want. For example, get_abstract gets us the article abstract, and get_cited_by returns the list of JSTOR articles citing a given article. The documentations below explain the specific purpose of each function in details. 

````python
def get_abstract(dom):
    """
    Given the dom of the summary page of a JSTOR article 
    (e.g. http://www.jstor.org/stable/info/25054401),
    return its abstract as a string
    """  
    
    abstract = dom.by_class("articleBody_abstractText")[0].content
    abstract = abstract.replace(',', '_')    
    # commas are replaced by '_' as the info is going to be stored in csv.
    
    return abstract
    
````

````python
def get_cited_by(dom):
    """
    Given the dom of the summary page of a JSTOR article 
    (e.g. http://www.jstor.org/stable/info/25054401),
    return a list of IDs of JSTOR articles that cite the given page
    """
    
    cited_by = []
   
    div_list = dom('div[id="itemsCiting"]')
    
    if len(div_list) == 1: #make sure there is a list of items citing this article
        div = div_list[0]
        for item in div('a[class="title"]'):
            # cited_by.append(item.href.split('/')[-1]) 
            # Revised by Xueshan. Explained in get_articles
            
            href = item.href                    
            loc = href.rfind('stable/')+len('stable/')  
            ID = href[loc:]
            cited_by.append(ID)            
    elif len(div_list) > 1:
        print 'Strange! More than one list of items citing this item at '+url
     
    return cited_by
````

````python
def get_articles(url):
    """
    Given the url of a result page on JSTOR Data for Research
    (e.g. http://dfr.jstor.org/?cs=ab%3A%28%27social+network%27%29^1.0&view=text&si=0),
    return a DataFrame with the following columns:
    
    ID: ID of the JSTOR article
    title: title of the JSTOR article
    journal: the journal where the article was published
    year: the year in which the article was published
    subjects: a list of subjects associated with the article, made into a string with subjects separated by '|'
    keywords: a list of keywords associated with the article, made into a string with keywords separated by '|'
    cited_by: a list of IDs of JSTOR articles citing the given article, made into a string with article IDs separated by '|'
    """
    
    data = defaultdict(list) #a dictoinary to store data
    dom = get_web(url)
    
    # Revised by Xueshan on Nov.28: 
    #get url_id, which can be used to check the completeness of the scraping.
    url_id_loc = url.rfind('=')
    url_id = int(url[url_id_loc+1:])
    
    for item in dom('ul[class="results_item"]'):
        title = item('a[class="title"]')[0]
        
        # Original line: ID = title.href.split('/')[-1] 
        # Revised by Xueshan on Nov. 27  
        # 
        # Most of the article IDs begin with 10.2307, 
        # with two valid url's 'http://www.jstor.org/stable/10.2307/210179' 
        # or 'http://www.jstor.org/stable/210179'
        # The url listed in "cited by" will be 'http://www.jstor.org/stable/210179'. 
        # Therefore, for these IDs, it is better to just store 210179.
        # 
        # However, some IDs are under the href 10.1086,
        # with url 'http://www.jstor.org/stable/10.1086/210179' both for information and cited-by.
        # I took 10.1086/210179 as the ID part for these articles.
        
        href = title.href                    
        loc = href.rfind('stable/')+len('stable/')  
        ID = href[loc:]
        if ID.split('/')[0] == '10.2307':
            ID = ID.split('/')[1]
        
        item_url = 'http://www.jstor.org/stable/info/%s'%ID
        item_dom = get_web(item_url)
         
            
        # Revised by Xueshan on Nov.28. Skip any articles with Errors, ex. IndexError.
        try:
            li_pub = item('li')[2] #the li tag containing the publication info
            journal = li_pub('cite')[0].content
            i = li_pub.content.find("), pp.")
            year = li_pub.content[i-4:i]
           
            li_sub = item('li')[3] #the li tag containing the subjects
            subjects = [HTMLParser().unescape(a.content) for a in li_sub('a[class="subject"]')]
            
            li_kw = item('li')[4] #the lig tag containing the keywords
            kws = [a.content for a in li_kw('a[class="keyterm"]')]
            
            citation = get_cited_by(item_dom)
            abstract = get_abstract(item_dom)
            
        except:
            print ""
            print "##############################################"
            print "Error at", ID
            print "##############################################"
            print ""
            logf = open('log_jump.txt', 'a')
            output = '%d, %s'%(url_id, ID)
            logf.write(output)
            logf.write('\n')
            logf.close()
            url_id += 1
            continue
        
        # Append article info to dataframe only when all the items are got from online successfully.
        data['id'].append(ID)
        
        data['title'].append(strip_tags(title.content).replace(',', '_'))   
        # commas are replaced by '_' as the info is going to be stored in csv.
        
        data['journal'].append(journal)
        data['year'].append(year)
        data['subjects'].append('|'.join(subjects))
        data['keywords'].append('|'.join(kws))
        data['cited_by'].append('|'.join(citation)) 
        data['abstract'].append(abstract)   # Added by Xueshan. Get the abstract
        
        data['url_id'].append(url_id)       
        # Added by Xueshan. Get the url_id. i.e. the trailing number in the url of search result.
         
        print url_id, ID, title.content
        
        url_id += 1
        
    return pd.DataFrame(data)
````

The following are the codes to get the subject classification scheme of journals from this webpage http://www.jstor.org/action/showJournals

Basically, we created a dictionary from the online classification scheme, such as the subject is the key, and the overarching discipline/section is the value. For example, African American Studies is grouped under Area Studies. We made each subject a separate key on purpose, so that later when we decide which discipline an article belongs to based on its primary subject, we can simply call dict_discipline_subject[subject].

````python
dom = get_web('http://www.jstor.org/action/showJournals')
sections, sublists = [], []

for section in dom('h3[class="sectionHeader"'):
    sections.append(section.content.split('(')[0].strip())
    
for sublist in dom.by_class('subListing'):
    temp = []
    
    for a in sublist('a'):
        temp.append(a.content.split('(')[0].strip())
        
    sublists.append(temp)
    

dict_discipline_subject = {} #make a dictionary

for sublist, section in zip(sublists, sections):
    for subject in sublist:
        #if the subject contains the html entity '&amp;', replace it with '&'
        #note by Xiaolin: I checked the list of subjects. there is only one html entity '&amp;', 
        #so the code does not need to handle any other html entities
        subject = subject.replace('&amp;', '&')
            
        dict_discipline_subject[subject] = section
        
#there are two subjects not in the JSTOR webpage, but in 
#Data for Research (where we got the network articles)
#add them manually here
dict_discipline_subject[u'Communication Studies'] = 'Social Sciences'
dict_discipline_subject[u'Developmental & Cell Biology'] = 'Science and Mathematics'
````

### Data Cleaning

This part explains how we cleaned and organized our data step by step. First, we created two csv files from articles_network.csv that contains the DataFrame of all articles: info.csv stores all the article info (i.e. the node attributes in our network), and cite.csv stores our edge data.

````python
datafile = 'articles_network.csv'
infofile = 'info.csv'
citefile = 'cite.csv'

infolist = ['id', 'journal', 'keywords', 'subjects', 'title', 'year', 'url_id', 'abstract']

# Extract header information.
data = open(datafile)
header = data.readline().strip().split(',')
data.close()

infoloc = [header.index(info) for info in infolist]
citeloc = header.index('cited_by')
idloc = header.index('id')

finfo = open(infofile, 'w')
fcite = open(citefile, 'w')

# Split data file into info and cite.
data = open(datafile)
for line in data:
    # Extract article information.
    line = line.strip().split(',')
    info = [line[loc] for loc in infoloc]
    output = ','.join(info)
    finfo.write(output)
    finfo.write('\n')
    
    # Extract network information. Stored in commonly-used network data format
    ID = line[idloc]
    cites = line[citeloc].strip().split('|')
    for cite in cites:
        output = ','.join([ID, cite])
        fcite.write(output)
        fcite.write('\n')

finfo.close()
fcite.close()
data.close()
````

### Step 2: Discipline-Level Network Structure

After data clean-up, the restricted data set contains 11841 citation ties, and this is the data set we are going to use to investigate the connectedness between disciplines. 

We first examine the directed citation network across disciplines in which each node represents a discipline. It is very important to make the direction of edges clear. A directed, weighted edge from discipline A to discipline B represents how many articles in discipline A are cited by the aricles in discipline B. The direction of the edges represents the direction of the flow of ideas. If the article A is cited by article B, then the idea in A flows to B. 

````python
def create_net(nodes, df):
    """
    This function is particularly created the citation network across disciplines.
    
    Given nodes (a list of node names in string) and df (a DataFrame of edge data),
    return the directed graph created from the given data. 
    
    Each node represents a discipline, and the edge weight represents the number of citations
    between the two disciplines.
    """
    
    graph = nx.DiGraph()
    
    #add nodes
    graph.add_nodes_from(nodes)
    
    #add edges and edge weights
    for (node1, node2) in it.permutations(nodes, 2):
        w = len(df[(df.from_dis==node1) & (df.to_dis==node2)])
        if w > 0:
            graph.add_edge(node1, node2, weight=w, difference=1.0/w)
            
    return graph
````

Here are the codes that determine the disciplines for the articles and then create the discipline citation network. 

````python
#in the edge data set, determine the discipline of both the 'from' article and the 'to' article
res_ties['from_dis'] = res_ties.id.apply(lambda i:valid_data[valid_data.id==i]['discipline'].iloc[0])
res_ties['to_dis'] = res_ties.cited_by.apply(lambda j:valid_data[valid_data.id==j]['discipline'].iloc[0])

#get all the nodes/disciplines
disciplines = valid_data.discipline.unique()

#create discipline citation network 
graph = create_net(disciplines, res_ties) 
````

Now we move on to network visualization. We re-used the codes of network visualization from homework 5. Since we will re-use the codes to create network graphs for the sub-network of each decade later, we made the codes into a function draw_graph. We visualize the discipline citation network with NODE SIZE representing the count of network articles in each discipline.

````python
def draw_graph(graph, pos_iterations, count_by_dis, title, k = None):
    """
    Given a discipline citation network, the number of position iterations to carry out,
    the number of articles in each discipline (used to display node size, a title for the graph, 
    and k (optimal distance between nodes), visualize the network.
    """
    
    plt.figure(figsize=(12, 7))
    
    #create a list of node size in the right order
    #in Networkx, if you create a list (L) of node display sizes of the same length as G.nodes(), 
    #then the size of G.nodes()[i] is set to be L[i] 
    node_size = [count_by_dis[d] for d in graph.nodes()]
    
    #create a list of edge weight labels in the right order
    edge_labels = dict([((u,v,),d['weight']) for u,v,d in graph.edges(data=True)])
    
    #this makes sure draw_spring results are the same at each call
    np.random.seed(1)  
    
    #determine position of each node using a spring layout
    pos = nx.spring_layout(graph, k=k, iterations=pos_iterations)
    
    #plot edges
    nx.draw_networkx_edges(graph, pos, alpha = 0.1)
    
    #plot nodes
    nx.draw_networkx_nodes(graph, pos, node_size=node_size, node_color = [0, 8, 2, 5, 4, 1, 6, 7, 3])
    
    #draw node labels
    lbls = nx.draw_networkx_labels(graph, pos, alpha=5, font_size=12)
    
    #draw edge weights as edge labels
    if edge_labels:
        nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
    
    #coordinate information is meaningless here, so let's remove it
    plt.xticks([])
    plt.yticks([])
    plt.title(title)
    remove_border(left=False, bottom=False)
````

The CCS captures the importance of the node in the network structure. The magnitude of the CCS measure relies on the connectedness of the node to other nodes in the network.
````python
def strip_non_ascii(string):
    """
    A helper function. 
    Returns the string without non ASCII characters
    """
    
    stripped = (c for c in string if 0 < ord(c) < 127)
    return ''.join(stripped)
````

````python
def create_article_net(node_df, tie_df):
    """
    Given a DataFrame of node attributes and a DataFrame of edge information,
    create a networkx.DiGraph from the data
    """
    
    graph = nx.DiGraph()
    
    #create a list of nodes in tuple format (node_id, dict(discipline=value_of_discipline))
    nodes = []
    for (index, row) in node_df.iterrows():
        nodes.append((row[0], dict(discipline=row[9], title = strip_non_ascii(row[4]))))
    
    graph.add_nodes_from(nodes)
    
    #create a list of edges in tuple format (from_node, to_node)
    edges = []
    for (index, row) in tie_df.iterrows():
        edges.append((row[0], row[1]))
    
    graph.add_edges_from(edges)
    
    return graph
````

````python
def draw_graph_art(graph, pos_iterations, title, k = None):
    """
    Visualize article citation network 
    """
    
    plt.figure(figsize=(25, 25))
    
    #create a list of node color in the right order
    
    nodedisp = nx.get_node_attributes(graph, 'discipline')
    nodecolor = [coldict[nodedisp[node]] for node in graph.nodes()]
    
    #this makes sure draw_spring results are the same at each call
    np.random.seed(1)  
    
    #determine position of each node using a spring layout
    pos = nx.spring_layout(graph, k=k, iterations=pos_iterations)
    
    #plot the edges
    nx.draw_networkx_edges(graph, pos, alpha = 0.02)
    
    #plot the nodes
    nx.draw_networkx_nodes(graph, pos, node_color = nodecolor)
    
    #draw the labels
    #lbls = nx.draw_networkx_labels(graph, pos, alpha=5, font_size=12)
    
    #coordinate information is meaningless here, so let's remove it
    plt.xticks([])
    plt.yticks([])
    plt.title(title)
    remove_border(left=False, bottom=False)
````

## A Case Study of *The Strength of Weak Ties*

Mark Granovetter's (1973) classic article *The Strength of Weak Ties* marked the start of interest in networks among sociologists and social scientists in general. Granovetter (1973) found that when people are looking for jobs, counterintuitively, their weak ties actually are better at providing novel job information and opportunities than their strong ties. The advantage of the weak ties is due to the fact that  individuals and their weak ties tend to be located in different and less-overlapping social networks, so the weak ties can bring in new information to the individuals. By contrast, individuals interact with their family members and friends frequently. They basically all know the same people and have access to the same kind of information. 

Granovetter's finding suggested a new way to look at the social world. Sociological research tends to focus on individual attributes as the explanations of social actions. For example, a woman who encounters glass ceiling in the workplace because of her gender. However, network researchers call the connections and social structure into our attention. In the view of network researchers, a woman is disadvantaged at work not simply because she is a woman, but due to the type of connections she has and the social relationships she is embedded in. She might have a larger proportion of kins in her networks and spend a lof of time socializing with kins and friends because of her reproductive role, so she misses the chance to develop ties that are not helpful for her career. 

Although in our dataset only about 800 articles cited *The Strength of Weak Ties* by 2010, the Google Scholar search shows that this article has been cited by 26,304 times. ## Step 1. Independent Diffusion Model

Modelling and forecasting the diffusion of innovations has been an academic interest since the 1960s and the after a few pioneers' work, Frank Bass developed a diffusion model in 1969 in his paper *A new product growth model for consumer durables* which quickly became the most popular model that describes the process of how new products get adopted in a population. The model is one of the most cited papers in marketing; up to Dec 2013 this paper has 5043 citations according to Google Scholar (accessed 12/12/2013). In 2004 it was selected as one of the ten most frequently cited papers in the 50-year history of Management Science and it was the only marketing paper on the list. 

Bass Diffusion Model is a ordinary differential equaition that describes the process of innovation adaptation. In our case, we will treat network science as a new innovation that get diffused in the academia community. Our model takes both the independent innovation dynamics and the "word of mouth" effect into account. The differential equation is

$$\frac{dN}{dt}(t)=p(m-N(t))+\frac{q}{m}N(t)(m-N(t))$$

where $N$ is the cumulative adopters, i.e. the total number of adopters so far. $p$ is the coefficient of the independent innovation dynamics, implying there is a certain percentage of unaffected researchers begin adopting this innovation at each time step. $q$ is the coefficient of the "word of mouth" effect, implying the adopters are spreading this innovation to potential adopters at each time step also. 

First we consider each discipline seperately. We use the *ode45* solver in Matlab to solve for the $N$ at each time, then call the *lsqnonlin* optimization solver to fit the parameters $m$, $p$, and $q$. For the starting times of each diffusion process, we take judicious guessing by inspecting the data.

First of all we define the Bass diffusion equation.

```matlab
function dN = bass_eqn(t,N,m,p,q)
dN = p*(m-N)+q/m*N*(m-N);
end
```

Then we set up the least square function

```matlab
function f = bass_lsq(x,data,t0)
m = x(1);
p = x(2);
q = x(3);
T = length(data);
for t = 1:t0
    f(t) = abs(0-data(t));
end 
for t = (t0+1):T
    [tt,NN] = ode45(@(tt,NN) bass_eqn(tt,NN,m,p,q), [t0,t], 0);
    f(t) = abs(data(t) - NN(length(NN)));
end
```

### Step 2. Cross-Disciplinary Diffusion Model

We introduce an cross-disciplinary diffusion model to quantify this effect. We add another term in the Bass diffusion differential equation, assuming this term is positively proportional to the number of in-degree of this discipline at this time and also positively proportional to the number of potential adopters. The new differential equation is 

$$\frac{dN}{dt}(t)=p(m-N(t))+\frac{q}{m}N(t)(m-N(t))+r*E(t)^{1.5}*(m-N(t))$$

where $E(t)$ is the number of in-degree of this discipline at this time, and $r$ is the coefficient of this term. This exponent would have actual meaning in physical diffusion problems, meaning the dimension of the diffusion space. However in our problem we have to arbitrarily set this exponent by a few trial and errors. We decided to use 1.5 becasue it fits the most data sets well. 

We need some additional information about the number of the in-degree of every discipline at each year. We calculate them, put them into dataframe and write into a csv file.

````python
weight_list = []
for year in range(1900,2011):
    ties = res_ties[(res_ties.from_year<=str(year)) & (res_ties.to_year<=str(year))]
    net = create_net(disciplines, ties)
    nodes = net.nodes()
    weights = {}
    for node in nodes:
        weights[node] = 0
    for i,j in net.edges():
        weights[j] = weights[j] + net[i][j]['weight']
    weight_list.append(weights)
    
weights = pd.DataFrame(weight_list)
weights.to_csv('weights.csv',sep=',',header=False,index=False)
````

First we write the cross-disciplinary diffusion equation.

```matlab
function dN = bass_eqn_cross(t,N,m,p,q,r,edges)
dN = p*(m-N)+q/m*N*(m-N)+r*edges^(1.5)*(m-N);
end
```

Then we write the cross-disciplinary diffusion least square fitting function

```matlab
function f = bass_lsq_cross(x,data,m,q,t0,edges)
p = x(1);
r = x(2);
T = length(data);
for t = 1:t0
    f(t) = abs(0-data(t));
end
for t = (t0+1):T
    [tt,NN] = ode45(@(tt,NN) bass_eqn_cross(tt,NN,m,p,q,r,edges(round(tt))), [t0,t], 0);
    f(t) = abs(data(t) - NN(length(NN)));
end
```

## V. Discussions

Network science has its root in computer science and discrete mathematics, and the many fundamental convergence theorems and approximation algorithms were explored by applied mathematician dates back to the early 1940s. However, only after the 1970s, the field of network science has formed and organized, and spread throughout the social sciences. Researchers quickly realized that the knowledge and techniques on the inter-node relations facilitates the analysis of individual behaviors and help understand the information diffusion process over the networks. This insight of the importance of network science has also sparked a rapid growth of development in medical science, natural science and business and economics. In fact, the expansion of network science in academia is just a sketch of the expansion of network science in our real life. The rapid development in telecommunication and internet technology strengthened the relation ties among people and this will eventually change the way how we live, work and think. 

The expanding process of network science itself has become a huge network now, and the data is particularly well-structured in academia because the citation network provides a data set with complete information of nodes, edges, and time stamps. We chose the keyword "network" to conduct our analysis, but our exploratory methods and quantitative models could easily be applied to any other rapidly growing, content rich and cross disciplinary sciences.

Our exploratory analysis revealed some very important features of the disciplines. As we summarized in the previous sections, some disciplines such as science and mathematics, social sciences, and business and economics are always the leading research fields and the ties between them are very strong. However the direction of the edges amongst them are not evenly distributed, for example, sociology has much more out-edges than in-edges while business and economics has much more in-edges than out-edges. This difference implies that in the field of network science, social science initiates more original ideas and business and economics adopts more methods from other disciplines and apply them to real world problems. Mathematics and natural science is simultaneously both creating new ideas and adopting ideas from other disciplines. Our case study on the paper *The Strength of Weak Ties* also verifies this result, as the original node is in sociology, then science and mathematics and business and economics quickly adopted this new idea and after a few citations a large number of business and economics paper emerged based on this idea.

The time stamps of the edges also contain very important information about the structure of the network. We can infer the information about flow of the development of network science amongst the disciplines from the time stamp data. One good approach could be the NetInf algorithm in Gomez-Rodriguez's paper *Inferring Networks of Diffusion and Influence*. This algorithm first estimates the structural parameters then simulates the diffusion process. This analysis is a good complement to our analysis on the edge information, as we didn't incorporate the time stamps into our model. Another worth mentioning method is the structural hole model. It looks for the optimal position on the network that provides the most marginal gain, i.e., the position that is most likely to be filled in the next time step. This analysis could provide useful prediction of the potential fertile topics for academic research.

Our diffusion model analysis gives a specific quantitative forecast for the future diffusion process. The core part of fitting the model to our data is nonlinear least square optimization. However nonlinear optimizations are notorious for its volatility so we fixed two parameters as constant throughout all disciplines, and their values are estimated by the weighted average of the preliminary optimization results. Another way to reduce the volatility of the model is to introduce hierarchical structural parameters, and assume the parameters of the 9 disciplines follow some specific distribution characterized by the hierarchical structural parameters.

The cross-disciplinary diffusion model is our biggest step forward from the previous research literatures. We introduced a new term representing the cross-disciplinary diffusion and look for a vector consists of two parameters that optimizes the residual norm function. This new model takes the number of in-edges into account and thus grows faster than our original model, and the growth rate is automatically adjusted with the number of in-edges, representing the information inflow of the discipline at this time. Our validation result showed that our new cross-disciplinary diffusion model provides a much more accurate prediction than our previous independent diffusion model.

The cross-disciplinary diffusion model can be further improved by adding dynamic control variables, and the model will become a constrained dynamic optimization problem. This model would be able to capture the delicate changes of the system and make optimized decision functions, answering the questions such as finding the best strategy of introduce a new technique to another discipline, and optimize the usage of a newly emerged research topic. However, constraint dynamic optimization often involves very complicated dynamic programming and has very high requirement of computational capacity.
